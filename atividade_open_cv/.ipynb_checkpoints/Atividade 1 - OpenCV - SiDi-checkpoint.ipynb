```python
media = np.vstack([
 np.hstack([ img1,      cv2.blur(img1, ( 3, 3))]),
 np.hstack([ cv2.blur(img1, ( 7, 7)), cv2.blur(img1, (9,9))]), 
])

plt.figure(figsize=(10,10))
plt.imshow(media, cmap="gray", vmin=0, vmax=255)
plt.title("Filtro da m√©dia")
plt.show()

#cv2.waitKey(0)
```

```python
img2 = cv2.imread("img2.jpg", 0)

plt.figure(figsize=(6,6))
plt.imshow(img2, cmap="gray", vmin=0, vmax=255)
plt.title("Filtro da m√©dia")
plt.show()
```

O melhor filtro para a imagem acima, seria o Mediana, pois segundo: https://projetos.extras.ufg.br/conpeex/2004/pibic/exatas/Vinicius.html
e https://seer.ufrgs.br/rita/article/view/rita_v16_n1_p125/7289
o filtro da Mediana √© mais inidicado para reduzir o ru√≠do "sal e pimenta", pois: "No filtro da mediana os valores dos pixels s√£o ordenados e o valor que ocupa a posi√ß√£o mediana √© selecionado para a posi√ß√£o (x, y) da imagem filtrada. Este filtro tende a reduzir o efeito de ru√≠do de pulso, do tipo "salt and peper", pois valores pontuais raramente aparecem juntos e portanto nunca ocupam a posi√ß√£o mediana."

```python
filtro_mediana = np.vstack([
 np.hstack([ img2,      cv2.medianBlur(img2, 3)])
])

plt.figure(figsize=(10,10))
plt.imshow(filtro_mediana, cmap="gray", vmin=0, vmax=255)
plt.title("Filtro da m√©dia")
plt.show()
```

```python
image5 = cv2.imread("img4.jpg", 1)
```

```python
plt.figure(figsize=(3,3))
#plt.imshow(image5)
#cv2.imshow("Original", image5)
#cv2.waitKey()
#plt.title("Original")
#plt.show()
```

```python
# c√≥digo de https://gist.github.com/FilipeChagasDev/bb63f46278ecb4ffe5429a84926ff812

def sepia_func(src_image):
    gray = cv2.cvtColor(src_image, cv2.COLOR_BGR2GRAY)
    normalized_gray = np.array(gray, np.float32)/255
    #solid color
    sepia = np.ones(src_image.shape)
    sepia[:,:,0] *= 153 #B
    sepia[:,:,1] *= 204 #G
    sepia[:,:,2] *= 255 #R
    #hadamard
    sepia[:,:,0] *= normalized_gray #B
    sepia[:,:,1] *= normalized_gray #G
    sepia[:,:,2] *= normalized_gray #R
    return np.array(sepia, np.uint8)
```

```python
sepia_img = sepia_func(image5)
```

```python
f_sepia = np.vstack([
 np.hstack([ image5, sepia_img])
])
```

```python
plt.figure(figsize=(10,10))
#plt.imshow(f_sepia)
cv2.imshow("Sepia", f_sepia)
cv2.waitKey(0)
cv2.destroyAllWindows() 
#plt.title("Filtro sepia")
#plt.show()
```

# Quest√£o 6

```python
image6 = cv2.imread("img5.jpg", 0)
```

**Binarizar a imagem**

```python
bins_edges_min_max = [0, 256]
num_bins = 256
bin_count, bins_edges = np.histogram(image6, num_bins, bins_edges_min_max)
bins_start = bins_edges[:-1]
```

```python
def draw_hist(x_axis, input):
    fig, ax = plt.subplots(figsize=(6,6))
    # why not using plt.hist? because we want to plot also some derivations of this hist, so this is easier
    plt.bar(x_axis, input, width=input.shape[0] / (x_axis[-1] - x_axis[0] + 1))
    return fig, ax


draw_hist(bins_start, bin_count)
plt.title("Original histogram")
plt.show()
```

```python
(T, bin_img6) = cv2.threshold(image6, 151, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(bin_img6, cmap="gray", vmin=0, vmax=255)
plt.title("Binariza√ß√£o")
plt.show()
```

**Dilata√ß√£o e eros√£o**

```python
cv2.imshow("Original", bin_img6)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
kernel = np.ones((5,5), np.uint8) 
  
img_erosion = cv2.erode(bin_img6, kernel, iterations=1)
img_dilation = cv2.dilate(img_erosion, kernel, iterations=1) 


cv2.imshow('Input', image6) 
cv2.imshow('Erosion', img_erosion) 
cv2.imshow('Dilation', img_dilation)
  
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
